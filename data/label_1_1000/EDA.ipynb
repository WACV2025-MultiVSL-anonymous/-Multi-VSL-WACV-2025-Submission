```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1->1000 train

```python
labels = pd.read_csv("train_1_1000_center_ord1.csv")
label_counts = labels.loc[labels['label'] < 1000]['label'].value_counts().sort_index()
num_subplots = 20
data_chunks = np.array_split(label_counts, num_subplots)
n_row = 10
n_col = 2
fig, axs = plt.subplots(num_subplots//n_col,n_col, figsize=(100,  60))
for i, data_chunk in enumerate(data_chunks):
    row = i // n_col
    col = i % n_col
    
    # V biu  ct
    data_chunk.plot(kind='bar', title=f'Biu  {i + 1}',ax=axs[row, col])
    axs[row, col].set_xlabel('Label')
    axs[row, col].set_ylabel('Count')
```

```python
import pandas as pd

# Danh sch cc file CSV bn mun hp nht
csv_files = [
    '/mnt/disk4/handsign_project/son_data/Experiment/data/label_1_400/test_1_400_center_ord1.csv',
    '/mnt/disk4/handsign_project/son_data/Experiment/data/label_1_400/test_1_400_left_ord1.csv',
    ]

# Danh sch lu tr cc DataFrame ch gm 2 ct 'name' v 'label'
dfs = []

# c tng file CSV, ch ly 2 ct 'name' v 'label' v thm vo danh sch DataFrame
for file in csv_files:
    df = pd.read_csv(file, usecols=['name', 'label'])
    dfs.append(df)

# Hp nht tt c cc DataFrame thnh mt DataFrame ln
merged_df = pd.concat(dfs, ignore_index=True)

# Lu DataFrame hp nht thnh file CSV mi
merged_df.to_csv('merged_file.csv', index=False)

print("CSV files have been successfully merged into merged_file.csv.")
```
